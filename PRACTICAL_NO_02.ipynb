```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# Sample data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
# Manual calculation of slope (m) and intercept (c)
n = len(X)
m = ((n * np.sum(X * y)) - (np.sum(X) * np.sum(y))) / ((n * np.sum(X**2)) - (np.sum(X)**2))
c = (np.sum(y) - m * np.sum(X)) / n
# Predictions using manual calculation
y_pred_manual = m * X + c
print(f"Manual Slope: {m}, Intercept: {c}")
# Using scikit-learn
model = LinearRegression()
model.fit(X.reshape(-1, 1), y)
y_pred_sklearn = model.predict(X.reshape(-1, 1))
print(f"Sklearn Coefficients: {model.coef_[0]}, Intercept: {model.intercept_}")
print("MSE:", mean_squared_error(y, y_pred_sklearn))
# Plot
plt.scatter(X, y, label="Data Points")
plt.plot(X, y_pred_sklearn, color='red', label="Regression Line")
plt.title("Simple Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
