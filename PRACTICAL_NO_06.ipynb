```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
import hdbscan
# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
# KMeans
kmeans = KMeans(n_clusters=4)
kmeans_labels = kmeans.fit_predict(X)

# DBSCAN
dbscan = DBSCAN(eps=1.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
# HDBSCAN
hdb = hdbscan.HDBSCAN(min_cluster_size=10)
hdb_labels = hdb.fit_predict(X)
# Evaluation
print("Silhouette KMeans:", silhouette_score(X, kmeans_labels))
print("Silhouette DBSCAN:", silhouette_score(X, dbscan_labels))
print("Silhouette HDBSCAN:", silhouette_score(X, hdb_labels))
# Visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, labels, title in zip(axes,
                              [kmeans_labels, dbscan_labels,
                               hdb_labels],
                              ['KMeans', 'DBSCAN', 'HDBSCAN']):
    ax.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    ax.set_title(title)
plt.show()
```
